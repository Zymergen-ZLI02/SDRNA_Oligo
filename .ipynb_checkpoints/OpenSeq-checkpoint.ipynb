```python
f = open("Compiled Seq.fasta","r")
seq = f.read()
allseqlist = seq.split(">")
print(allseqlist)
```

```python
myseqdict = {}

for s in range(1,len(allseqlist)):
    
    splitlist = allseqlist[s].split()
    seqstart = len(splitlist[0])+len(splitlist[1])+len(splitlist[2])+len(splitlist[3])+3
    seqreturn = allseqlist[1][seqstart:]
    seq = seqreturn.replace('\n','')

    myseqdict.update({splitlist[0]:seq})
    
```

```python
print(myseqdict)
```
